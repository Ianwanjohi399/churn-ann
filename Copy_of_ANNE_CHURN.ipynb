```python
# Artificial Neural Network for Customer Churn

# Step 1 - Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 2 - Upload dataset in Colab
from google.colab import files

print("Please upload your dataset (Churn_Modelling.csv)")
uploaded = files.upload()  # Choose the file from your computer

# Get the uploaded filename
filename = list(uploaded.keys())[0]
print("Uploaded file:", filename)

# Step 3 - Load dataset
dataset = pd.read_csv(filename)
print("Dataset shape:", dataset.shape)
print(dataset.head())

# Step 4 - Select features and target
X = dataset.iloc[:, 3:13].values   # Features (CreditScore → EstimatedSalary)
y = dataset.iloc[:, 13].values     # Target (Exited column)

# Step 5 - Encode categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Encode Geography
labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])

# Encode Gender
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])

# OneHotEncode Geography
columnTransformer = ColumnTransformer(
    [('encoder', OneHotEncoder(), [1])],
    remainder='passthrough'
)
X = np.array(columnTransformer.fit_transform(X), dtype=np.float64)

# Avoid dummy variable trap (drop 1 dummy variable)
X = X[:, 1:]

# Step 6 - Split dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)

# Step 7 - Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Step 8 - Build the ANN
from keras.models import Sequential
from keras.layers import Dense

classifier = Sequential()

# Input + first hidden layer
classifier.add(Dense(units=6, kernel_initializer='uniform',
                     activation='relu', input_dim=11))

# Second hidden layer
classifier.add(Dense(units=6, kernel_initializer='uniform',
                     activation='relu'))

# Output layer
classifier.add(Dense(units=1, kernel_initializer='uniform',
                     activation='sigmoid'))

# Compile ANN
classifier.compile(optimizer='adam',
                   loss='binary_crossentropy',
                   metrics=['accuracy'])

# Step 9 - Train the ANN
classifier.fit(X_train, y_train, batch_size=10, epochs=100)

# Step 10 - Evaluate model
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)

print("Confusion Matrix:\n", cm)
print("Accuracy:", acc)

# Step 11 - Predict single customer
"""
Geography: France
Credit Score: 600
Gender: Male
Age: 40
Tenure: 3
Balance: 60000
Number of Products: 2
Has Credit Card: Yes
Is Active Member: Yes
Estimated Salary: 50000
"""

new_prediction = classifier.predict(
    sc.transform(np.array([[0.0, 0, 600, 1, 40, 3,
                            60000, 2, 1, 1, 50000]]))
)
print("Single customer churn prediction:", new_prediction > 0.5)
```
